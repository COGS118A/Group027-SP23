One evaluation metric that can be used to quantify the performance of both the benchmark model and the solution model in the context of housing price prediction is Mean Squared Error (MSE).
MSE calculates the average squared difference between the predicted values and the actual values. It measures the average magnitude of the error, with larger errors being penalized more heavily. 

MSE is derived by taking the average of the squared differences between the predicted and actual values:

\begin{align*}
MSE = \sum_{i=1}^{n}(y_i - \hat{y}_i)^2
\end{align*}

MSE is widely used in regression problems and provides a measure of the average prediction error. For example if given the following predictions from a model:
| Acutal Price ($y$) | Predicted Price  ($\hat{y}$)|
|---|---|
| 150000       | 155000          |
| 180000       | 175000 |
| 200000 | 195000|
|220000 | 230000|
|250000 | 240000|

$MSE = \frac{(155,000−150,000)^2 + (175,000−180,000)^2 + (195,000−200,000)^2 + (230,000−220,000)^2 + (240,000−250,000)^2}{5} = 175,000,000.$ 

This would mean that, on average, the predicted value of housing prices deviates from the the actual price by $\sqrt{175,000,000} = 175,000$ dollars. We can use this evaluation metric to compare the performance of various models where a smaller MSE would typically represent better performance.

```python
house_data = house_data.drop(['condition', 'grade', 'lat', 'long'], axis=1)
house_data.shape
```

By plotting a heatmap, we can identify which features are strongly correlated with each other and with the target variable, "price" in this case. This analysis helps us understand the interdependencies between variables and provides valuable information for feature selection, identifying important predictors, and making informed decisions during the modeling process. Let's dive into generating a heatmap to explore the correlations within our house_data and gain valuable insights into the dataset.

```python
corr_matrix = house_data.corr()

# Generate a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Map")
plt.show()
```

One of the notable observations from the heatmap analysis is that the "price" variable in our house_data dataset exhibits a strong positive correlation with the "sqft_living" feature. Lets keep this in mind as we construct our models. To start, we need to seperate the date into X and y and then further splitting into a training and test set. We should also scale the date in order to prevent some features from dominating the learning process solely based on their natural larger values.

```python
X = house_data.drop('price', axis=1)  
y = house_data['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
scaler = StandardScaler()
```

```python
#Scale the features 
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Now that we have split the data, we can generate base models for various different machine learning algorithms. We start by creating svm, linear regression, random forest, and gradient boost ensemble models of our training data and use each model to make a prediction on our test data.

```python
#Make a prediction using SVM
svm = SVR(kernel='linear')
svm.fit(X_train_scaled, y_train)
svm_y_pred = svm.predict(X_test_scaled)
```

```python
#Make a prediction using linear regression
linear_regression = LinearRegression()
linear_regression.fit(X_train_scaled, y_train)
lr_y_pred = linear_regression.predict(X_test_scaled)
```

```python
#Make a prediction using Random Forest
random_forest = RandomForestRegressor()
random_forest.fit(X_train_scaled, y_train)
rf_y_pred = random_forest.predict(X_test_scaled)
```

```python
#Make a prediction using Gradient Boosting
gradient_boosting = GradientBoostingRegressor()
gradient_boosting.fit(X_train_scaled, y_train)
gb_y_pred = gradient_boosting.predict(X_test_scaled)
```

Now that we have the predictions made by each model, we can calculate the mean squared error of each base model and compare the performance between the different machine learnign algorithms.

```python
sv_mse = mean_squared_error(y_test, svm_y_pred)
print("Mean Squared Error for SVM:", sv_mse)
lr_mse = mean_squared_error(y_test, lr_y_pred)
print("Mean Squared Error for Linear Regression:", lr_mse)
rf_mse = mean_squared_error(y_test, rf_y_pred)
print("Mean Squared Error for Random Forests:", rf_mse)
gb_mse = mean_squared_error(y_test, gb_y_pred)
print("Mean Squared Error for Gradient Boosting:", gb_mse)
```

After calculating the mean squared error for each model, it would appear that gradient boosting performs the best compared to the other models. Let's perform cross validation to confirm if this actually is the case. We partition the data into 5 folds and get the cross validation score of each model where each model is scored by its mean squared error. The model with the lost average mean squared error should theoretically have the best performance.

```python
X_scaled = scaler.fit_transform(X)
models = [('SVM', svm), ('Random Forest', random_forest), ('Linear Regression', linear_regression), ('Gradient Boosting', gradient_boosting)]
for name, model in models:
    scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
    mse_scores = -scores
    avg_mse = mse_scores.mean()
    print(f'{name}: Average MSE = {avg_mse}')
```

Upon performing cross validation, we can confirm that the Gradient Boosted Ensemble Regressor is our best model. Next, lets use grid search to perform some hyperparameter tunring. We first need to define our parameter grid. Then, we can perform a grid search of this parameter grid to determine the best parameters for our gradient boosted model. 

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.05, 0.01],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'subsample': [0.8, 1.0],
}

# Perform grid search
grid_search = GridSearchCV(estimator=gradient_boosting, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_scaled, y)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = -grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score (MSE):", best_score)
```

Performing the grid search has determined the above hyperparameters to be the best. The mean squared error for a gradient boosted model using these hyperparameters is also lower compared to any of the other base models. Now that we have determined the best hyperparameters for our model, lets generate a new Gradient Boosted Ensemble using these hyperparameter settings.

```python
gb_best = GradientBoostingRegressor(**best_params)

# Fit the model to the scaled features and target variable
gb_best.fit(X_train_scaled, y_train)
gb_b_pred = gb_best.predict(X_test_scaled)
gb_b_mse = mean_squared_error(y_test, gb_b_pred)
gb_b_mse
```

We can now analyize the the importance of each feature within our model and compare these results with the heatmap generated earlier. Lets plot the importance of each feature within our dataset in a bar chart.

```python
feature_importances = gb_best.feature_importances_

# Sort feature importances in descending order
sorted_indices = np.argsort(feature_importances)[::-1]
sorted_feature_importances = feature_importances[sorted_indices]

# Get the names of the features
feature_names = X.columns[sorted_indices]

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_names)), sorted_feature_importances)
plt.xticks(range(len(feature_names)), feature_names, rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance Plot')
plt.tight_layout()
plt.show()
```

As we can see, sqft_living is clearly the most important feature. This makes sense as it was seen to have the highest correlation with price from the heatmap. Lets try to improve our model using nested cross-validation. Two nested cross-validation loops are used: the outer loop and the inner loop. Within each fold of the outer loop, a grid search is performed. The estimator parameter is set to the GBR model, the param_grid is specified, and the cv parameter is set to the inner loop (3-fold cross-validation in this case). The grid search finds the best hyperparameters based on the evaluation scores. After the grid search is complete, the best model is obtained using the best_estimator_ attribute of the GridSearchCV object. The model is then evaluated on the validation set (test set for each fold of the outer loop) using the mean squared error. The best model is selected based on the lowest MSE score across all folds of the outer loop and trained on the entire dataset. The performance of the final model is evaluated on the test set using the mean squared error

```python
# Define the parameter grid for hyperparameter tuning
param_grid = {
    'learning_rate': [0.1, 0.01],
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'subsample': [0.8, 1.0]
}

# Initialize the GBR model
gbr = GradientBoostingRegressor()

# Perform nested cross-validation
outer_cv = KFold(n_splits=5, shuffle=True, random_state=42)
inner_cv = KFold(n_splits=3, shuffle=True, random_state=42)

grid_search = GridSearchCV(estimator=gbr, param_grid=param_grid, cv=inner_cv)
best_model = None
best_score = float('inf')

for outer_train_index, outer_test_index in outer_cv.split(X):
    X_train, X_test = X_scaled[outer_train_index], X_scaled[outer_test_index]
    y_train, y_test = y[outer_train_index], y[outer_test_index]

    # Perform hyperparameter tuning within each fold of the outer loop
    grid_search.fit(X_train, y_train)

    # Get the best model and evaluate it on the validation set
    model = grid_search.best_estimator_
    score = mean_squared_error(y_test, model.predict(X_test))

    # Check if this model performs better than the previous best
    if score < best_score:
        best_model = model
        best_score = score

# Train the final model on the entire dataset
best_model.fit(X_scaled, y)

# Evaluate the final model on the test set
final_score = mean_squared_error(y_test, best_model.predict(X_test))
```

```python
print(f' Mean Squared Error of Best Model = {final_score}')
```

Generating the bar chart for the feature importance of our new best model, we can see that this new model has value sqft_living even more compared to the previous one.

```python
feature_importances = best_model.feature_importances_

# Sort feature importances in descending order
sorted_indices = np.argsort(feature_importances)[::-1]
sorted_feature_importances = feature_importances[sorted_indices]

# Get the names of the features
feature_names = X.columns[sorted_indices]

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_names)), sorted_feature_importances)
plt.xticks(range(len(feature_names)), feature_names, rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance Plot')
plt.tight_layout()
plt.show()
```

Now fitting this model to the same data using a train/test split setup again, we can see that the mean squared error of this model is even lower than the previous models and substantially much smaller than the average MSE of the base models.

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
best_model.fit(X_train, y_train)
y_pred_test = best_model.predict(X_test)
test_score = mean_squared_error(y_test, y_pred_test)
test_score
```
